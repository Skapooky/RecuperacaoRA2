# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 2.
$$h(x) = \frac{\ln(x)}{x}$$

1

Derivando separadamente cada elemento:

$$\frac{d}{dx}(x^3) = 3x^2$$

$$\frac{d}{dx}(5x^2) = 2\times5x^2$$
$$\frac{d}{dx}(5x^2) = 10x^2$$

$$\frac{d}{dx}(-7x)$$
$$-7$$

$$\frac{d}{dx}(9)$$
$$0$$

Somando tudo:
$$\frac{d}{dx}(x^3 + 5x^2 - 7x + 9) = 3x^2 + 10x -7$$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x**3 + 5*x**2 - 7*x + 9


x = np.linspace(-8, 10, 1000)
y = f(x)

plt.plot(x, y, label=r'$f(x) = x^3 + 5x^2 - 7x + 9$')
plt.legend()
plt.title("Função")
plt.grid()
plt.show()
```

```python
def d(x):
  return 3*x**2 + 10*x -7


x = np.linspace(-10, 10, 1000)
y = d(x)

plt.plot(x, y, label=r'$d(x) = 3x^2 + 10x -7$')
plt.legend()
plt.title("Derivada")
plt.grid()
plt.show()
```

2.

Regra do quociente
$$\frac{d}{dx}(\frac{\ln(x)}{x}) = \frac{\frac{d}{dx}(\ln(x))x - \ln(x)\frac{d}{dx}(x)}{x^2}$$
$$\frac{d}{dx}(\frac{\ln(x)}{x}) = \frac{\frac{1}{x}x - \ln(x)1}{x^2}$$
$$\frac{d}{dx}(\frac{\ln(x)}{x}) = \frac{\frac{x}{x} - \ln(x)}{x^2}$$
$$\frac{d}{dx}(\frac{\ln(x)}{x}) = \frac{1 - \ln(x)}{x^2}$$

```python
def f(x):
  return np.log(x) / x

x = np.linspace(-1, 1, 1000)
y = f(x)

plt.plot(x, y, label=r'$f(x) = \frac{\ln(x)}{x}$')
plt.legend()
plt.title("Função")
plt.grid()
plt.show()
```

```python
def d(x):
  return (1 - np.log(x)) / (x**2)

x = np.linspace(-1, 1, 1000)
y = d(x)

plt.plot(x, y, label=r'$f(x) = \frac{1 - \ln(x)}{x^2}$')
plt.legend()
plt.title("Derivada")
plt.grid()
plt.show()
```

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
x = sp.Symbol('x')
fx = x**2 - 98*x + 4
derivadaX = sp.diff(fx, x)

pontos_criticos = sp.solve(derivadaX, x)


print(f"A função não possui máximo local visto que é ela não possui nenhum pico, e seu único ponto crítico é {pontos_criticos[0].evalf():.2f}:")


def f(x):
  return x**2 - 98*x + 4

x = np.linspace(0, 100, 1000)
y = f(x)

plt.plot(x, y)
plt.grid()
plt.show()
```
